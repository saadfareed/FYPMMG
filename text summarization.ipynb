### 5. Compute Word Frequency


```python
# Build Word Frequency
# word.text is tokenization in spacy
word_frequencies = {}
for word in docx:
    if word.text not in stopwords:
         if word.text not in punctuation:
            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
```

```python
print(word_frequencies)
```

```python
# Maximum Word Frequency
maximum_frequency = max(word_frequencies.values())
```

```python
maximum_frequency
```

```python
#normalize
for word in word_frequencies.keys():  
        word_frequencies[word] = (word_frequencies[word]/maximum_frequency)
```

```python
# Frequency Table
print(word_frequencies)
```

```python
# Sentence Tokens
sentence_tokens= [ sent for sent in docx.sents ]
print(sentence_tokens)
```

### 6. Compute Sentence  Scores


```python
# Sentence Score via comparrng each word with sentence
sentence_scores = {}  
for sent in sentence_tokens:  
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sentence_scores.keys():
                        sentence_scores[sent] = word_frequencies[word.text.lower()]
                else:
                        sentence_scores[sent] += word_frequencies[word.text.lower()]
```

#### Finding Top N Sentence with largest score

```python
from heapq import nlargest
```

```python
select_length=int(len(sentence_tokens)*0.2)
select_length
```

### 7. Build summary 


```python
summarized_sentences = nlargest(select_length, sentence_scores, key=sentence_scores.get)
```
