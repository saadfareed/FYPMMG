# Length Score Summarization

```python
import numpy
final = []
file=open("BBC.txt",'r')
inputText=file.read()
sentence=inputText
sentlen=len(sentence)
saad=sentence.split(".")
#print(len(saad))
#print("\n")
result=[]
def sentencelen(saad):
    for i in range(len(saad)):
        ilen=len(saad[i])
        #print(ilen)
        alpha=ilen**0.5
        #print(alpha)
        if(alpha==0):
            continue
        beta=1/alpha
        results=final.append(beta)
        #print("\n")
    maximum=max(final)
    #print(maximum)
    for i in final:
        gema=i/maximum
        #print(gema)
        res=result.append(gema)
    return 0
sentencelen(saad)

        
sort_index = numpy.argsort(result)[::-1]
#print(sort_index)
sort_index=list(sort_index)
length = len(sort_index)
#print(length)
sum_index = length//3
#print(sum_index)
summary = sort_index[:sum_index]
#print(summary)
print("Summary...")
for i in summary:
    print(saad[i])

#print("\n")
#print(saad[1])
```

```python
import json
import re
import requests
accessToken = "bf8426d5-7e47-4f3d-9802-fc827f74cfb5"
file=open("BBC.txt",'r')
frequency={}
sadi={}
inputText=file.read()
JSON_MSG = "{ \"text\" : \"" + inputText + "\" , \"token\" : \"" + accessToken + "\" }"
#print(JSON_MSG)
response = requests.post('https://api.cle.org.pk/v1/pos',json={"text":inputText,"token":accessToken})
response=response.json()
with open('pos.json','w') as f:
    json.dump(response,f)
f.close()
with open('pos.json',"r") as w:
    saad=json.load(w)
w.close()
#print(type(saad))
fareed=saad['response']
sadi=(fareed['tagged_text'])
#print(sadi)
import numpy
sentence_li=[]
x = sadi.split(".")
DF={}
#print(x)
for i in range(len(x)):
    #print(x[i])
    sentence_li.append(x[i])
   # print("\n")
#print(sentence_li)
#print(len(sentence_li))

NN_li=[]
NNP_li=[]
other_li=[]
for i in range(len(sentence_li)):
    sentence=str(sentence_li[i])
    s=sentence.split(" ")
    NN=0
    NNP=0
    other=0
    for w in s:
        f=str(w)
        f=f.split("|")
        #print(f)
        for w in f[::2]:
            #print(w)
            if(w=="NN"):
                NN=NN+1
            elif(w=="NNP"):
                NNP=NNP+1
            else:
                other=other+1
      
    NN_li.append(NN)
    NNP_li.append(NNP)
    other_li.append(other)
    
print("List of Noun Singular in sentence: ")
print(NN_li)
print("\n")
print("List of Proper Noun Singular in sentence: ")
print(NNP_li)
print("\n")
result=NN_li
for i in range(len(result)):
    result[i]=result[i]+NNP_li[i]
    
print("Resultant Score list of sentence: ")
print(result)
print("\n")
fi=open("BBC.txt",'r')
textfile=fi.read()
tex=textfile.split(".")
sort_index = numpy.argsort(result)[::-1]
#print(sort_index)
sort_index=list(sort_index)
length = len(sort_index)
sum_index = length//3
summary = sort_index[:sum_index]
#print(summary)
print("\n")
print("Summary...\n")
for i in summary:
    print(tex[i])
```
